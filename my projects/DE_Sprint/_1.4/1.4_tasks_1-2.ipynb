<h2> Задание 1. Построение распределений </h2>

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import rayleigh
```

```python
# сгенерируем выборку с распределением Рэлэя
r = np.random.rayleigh(size = 10000)
r
```

```python
# построим ее график c помощью библиотеки seaborn
sns.displot(r, kde=True, stat="density", linewidth=0)

plt.show()
```

```python
# построим ее график с помощью библиотеки matplotlib
plt.figure()
plt.hist(r, bins = 50, density = True,  alpha = 0.4, linewidth=0.5)
plt.show()
```

<h2> Задание 2. Построение графика на по функции </h2>

```python
# дана функция (x*2 + y*2 - 1)*3 - x*2*y*3 = 0
# необходимо построить ее график


# найдем корни уравнения относительно x, при которых D=0, и ограничим ими наши значения x
x1 = -1.139028
x2 = 1.139028

# создадим массив значений x с маленьким шагом
x = np.linspace(x1, x2, 100)
```

```python
# поситаем значения y 
y1 = 1/2*((x**2)**(1/3)+((x**4)**(1/3)-4*x**2+4)**1/2)
y2 = 1/2*((x**2)**(1/3)-((x**4)**(1/3)-4*x**2+4)**1/2)
```

```python
#xx = np.concatenate([x,x])
#yy = np.concatenate([y1,y2])
```

```python
# построим получившиеся зависимости

fig, ax = plt.subplots()

plt.plot(x,y2)
plt.plot(x,y1)

plt.show()

```
